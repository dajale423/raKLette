```python
chrom = "*"

filename = os.path.join(KL_data_dir, f"results/footprints/zscore/AC/{chrom}.tsv")
ddf = dd.read_csv(filename, sep = "\t", dtype={'Spliceai_info': 'object'})
df_ac = ddf.compute()
```

```python
df_ac["zscore_ac"] = (df_ac["AC_total"] - df_ac["expected_AC"])/np.sqrt(df_ac["AC_var"])
```

## after filtering out common variants (MAF < 0.05)

```python
## histogram of expected allele count
sns.histplot(data=df_ac, x="expected_AC")
```

```python
## distribution of mu is already skewed --> maybe we should limit the mu

sns.histplot(data=df_ac, x="mu")
```

```python
print(df_ac["mu"].quantile(q=0.1))
print(df_ac["mu"].quantile(q=0.4))
print(df_ac["mu"].quantile(q=0.5))
print(df_ac["mu"].quantile(q=0.6))
print(df_ac["mu"].quantile(q=0.9))
```

```python
## filter for mu for quantile 0.4 < q < 0.6

upper_quantile = 0.41
lower_quantile = 0.39

lower_boundary = df_ac["mu"].quantile(q=lower_quantile)
print(lower_boundary)

upper_boundary = df_ac["mu"].quantile(q=upper_quantile)
print(upper_boundary)

df_ac_mu = df_ac[(df_ac["mu"] > lower_boundary) & (df_ac["mu"] < upper_boundary)]

len(df_ac_mu)
```

```python
## histogram of expected allele count
sns.histplot(data=df_ac_mu, x="mu")
```

```python
sns.histplot(data=df_ac_mu, x="expected_AC")
```

```python
from scipy.stats import norm

# Plot the histogram with density
sns.histplot(data=df_ac, x="zscore_ac", kde=True, stat = "density")

# Overlay the standard normal distribution
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), 'r-', lw=2, label='Standard Normal Distribution')
plt.xlim(-10,10)

# Add legend and labels
plt.legend()
plt.xlabel('Z-score')
plt.ylabel('Density')

# Show the plot
plt.show()
```
